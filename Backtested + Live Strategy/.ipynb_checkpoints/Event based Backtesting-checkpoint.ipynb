```python
import pandas as pd
import telegram_bot
import datetime
import talib
from glob import glob
```

```python
data = pd.read_csv('../Practice Projects/2020_data/MERGED FILES/BHARTIARTL_F1.csv')
data['date'] = data['date'].apply(str)
data['datetime'] = data['date'] + data['time']
data['date'] = data['date'].apply(lambda x : datetime.datetime.strptime(x, '%Y%m%d'))
data['time'] = data['time'].apply(lambda x : datetime.datetime.strptime(x, '%H:%M').time())
data['datetime'] = data['datetime'].apply(lambda x : datetime.datetime.strptime(x, '%Y%m%d%H:%M'))
data = data[['datetime','ticker','open','high','low','close','volume','date','time']]
data.set_index('datetime', inplace=True)
```

```python
folder = '../Practice Projects/2020_data/MERGED FILES/*'
```

```python
destination = '../Practice Projects/2020_data/PICKLE FILES/'
```

```python
for file in glob(folder):
    instrument = file.split('/')[-1].split('.')[0] + '.pkl'
    data = pd.read_csv(file)
    data['date'] = data['date'].apply(str)
    data['datetime'] = data['date'] + data['time']
    data['date'] = data['date'].apply(lambda x : datetime.datetime.strptime(x, '%Y%m%d'))
    data['time'] = data['time'].apply(lambda x : datetime.datetime.strptime(x, '%H:%M').time())
    data['datetime'] = data['datetime'].apply(lambda x : datetime.datetime.strptime(x, '%Y%m%d%H:%M'))
    data = data[['datetime','ticker','open','high','low','close','volume','date','time']]
    data.set_index('datetime', inplace=True)
    data.to_pickle(destination + instrument)
```

```python
def price_range(current_price, price_reference, percentage):

    return ((price_reference * (1 - percentage/100) < current_price) & 
                (current_price < price_reference * (1 + percentage/100)))
```

```python
all_dates = list(pd.Series(data.index).apply(lambda x : x.date()).unique())[2:]
capital = 100000
percentage_padding = 0.02
position_size = 150
stoploss = 10

for today in pd.to_datetime(all_dates):

    df = data[(data['date'] > today - datetime.timedelta(days=3))
              & (data['date'] < today)]

    candle = pd.DataFrame()

    prices_calculated_flag = False

    long_position_flag = False
    long_target_1_flag = False
    long_target_2_flag = False
    short_position_flag = False
    short_target_1_flag = False
    short_target_2_flag = False

    for index, row in df.iterrows():

        if ((index.date() == today) &
            (index.time() >= datetime.time(hour=9, minute=15)) &
            (index.time() <= datetime.time(hour=9, minute=30))):

            candle = candle.append(row)
            df = df.append(row)

        if ((index.date() == today) & 
           (index.time() > datetime.time(hour=9, minute=30)) & 
           (prices_calculated_flag == False)):

            high = candle['high'].max()
            low = candle['low'].min()

            diff = round(high - low, 3)
            atp = round((high + low) / 2, 3)

            derived = round(diff * 0.618, 3)
            stoploss = stoploss

            buy = round(atp + stoploss, 3)
            sell = round(atp - stoploss, 3)

            buy_momentum = round(atp + derived, 3)
            sell_momentum = round(atp - derived, 3)

            long_target_1 = buy_momentum + diff
            long_target_2 = long_target_1 + diff

            short_target_1 = sell_momentum - diff
            short_target_2 = short_target_1 - diff

            prices_calculated_flag = True

        if ((index.date() == today) &
            (index.time() > datetime.time(hour=9, minute=30))):

            df = df.append(row)
            df['cci'] = talib.CCI(df['high'],
                                  df['low'],
                                  df['close'],
                                  timeperiod=369)

            ltp = df.iloc[-1]['close']
            cci = df.iloc[-1]['cci']

            if (long_position_flag | short_position_flag) == False:

                if price_range(ltp, buy, percentage_padding) & (cci > 20):

                    long_position_flag = True
                    long_target_1_flag = False
                    long_target_2_flag = False
                    capital = capital - (position_size * ltp)
                    #telegram_bot.send(f'LONG ENTRY - Price : {ltp} \nGameplan Buy Price : {buy}')
                    #print(f'{index} \nLONG ENTRY - Price : {ltp} \nGameplan Buy Price : {buy} \nCapital : {capital}')
                    print(f'{index} \nLONG ENTRY - Price : {ltp} \nGameplan Buy Price : {buy}')
                    print('______________________________________________')
                    
                elif price_range(ltp, sell, percentage_padding) & (cci < -20):

                    short_position_flag = True
                    short_target_1_flag = False
                    short_target_2_flag = False
                    capital = capital + (position_size * ltp)
                    #telegram_bot.send(f'LONG ENTRY - Price : {ltp} \nGameplan Sell Price : {sell}')
                    #print(f'{index} \nSHORT ENTRY - Price : {ltp} \nGameplan Sell Price : {sell} \nCapital : {capital}')
                    print(f'{index} \nSHORT ENTRY - Price : {ltp} \nGameplan Sell Price : {sell}')
                    print('______________________________________________')
        
            if long_position_flag:

                if (ltp > long_target_1) & (long_target_1_flag == False):

                    long_target_1_flag = True
                    capital = capital + (position_size/2 * ltp)
                    #telegram_bot.send(f'LONG TARGET 1 - Price : {ltp} \nGameplan Target 1 Price : {long_target_1}')
                    #print(f'{index} \nLONG TARGET 1 - Price : {ltp} \nGameplan Target 1 Price : {long_target_1} \nCapital : {capital}')
                    print(f'{index} \nLONG TARGET 1 - Price : {ltp} \nGameplan Target 1 Price : {long_target_1}')
                    print('______________________________________________')

                elif (ltp > long_target_2) & (long_target_2_flag == False):

                    long_target_2_flag = True
                    long_position_flag = False
                    capital = capital + (position_size/2 * ltp)
                    #telegram_bot.send(f'LONG TARGET 2 - Price : {ltp} \nGameplan Target 2 Price : {long_target_2}')
                    print(f'{index} \nLONG TARGET 2 - Price : {ltp} \nGameplan Target 2 Price : {long_target_2} \nCapital : {capital}')
                    print('______________________________________________')

                elif ltp < sell:

                    if long_target_1_flag:

                        long_position_flag = False
                        capital = capital + (position_size/2 * ltp)
                        #telegram_bot.send(f'STOP LOSS - Price : {ltp} \nGameplan Sell Price : {sell}')
                        #print(f'{index} \nSTOP LOSS - Price : {ltp} \nGameplan Sell Price : {sell} \nCapital : {capital}')
                        print(f'{index} \nSTOP LOSS - Price : {ltp} \nGameplan Sell Price : {sell}')
                        print('______________________________________________')

                    else:

                        long_position_flag = False
                        capital = capital + (position_size * ltp)
                        #telegram_bot.send(f'STOP LOSS - Price : {ltp} \nGameplan Sell Price : {sell}')
                        print(f'{index} \nSTOP LOSS - Price : {ltp} \nGameplan Sell Price : {sell} \nCapital : {capital}')
                        print('______________________________________________')

                    if cci < -20:

                        short_position_flag = True
                        capital = capital + (position_size * ltp)
                        #telegram_bot.send(f'SHORT ENTRY - Price : {ltp} \nGameplan Sell Price : {sell}')
                        #print(f'{index} \nSHORT ENTRY - Price : {ltp} \nGameplan Buy Price : {sell} \nCapital : {capital}')
                        print(f'{index} \nSHORT ENTRY - Price : {ltp} \nGameplan Buy Price : {sell}')
                        print('______________________________________________')

            if short_position_flag:

                if (ltp < short_target_1) & (short_target_1_flag == False):

                    short_target_1_flag = True
                    capital = capital - (position_size/2 * ltp)
                    #telegram_bot.send(f'SHORT TARGET 1 - Price : {ltp} \nGameplan Target 1 Price : {short_target_1}')
                    #print(f'{index} \nSHORT TARGET 1 - Price : {ltp} \nGameplan Target 1 Price : {short_target_1} \nCapital : {capital}')
                    print(f'{index} \nSHORT TARGET 1 - Price : {ltp} \nGameplan Target 1 Price : {short_target_1}')
                    print('______________________________________________')

                elif (ltp < short_target_2) & (short_target_2_flag == False):

                    short_target_2_flag = True
                    short_position_flag = False
                    capital = capital - (position_size/2 * ltp)
                    #telegram_bot.send(f'SHORT TARGET 2 - Price : {ltp} \nGameplan Target 2 Price : {short_target_2}')
                    print(f'{index} \nSHORT TARGET 2 - Price : {ltp} \nGameplan Target 2 Price : {short_target_2} \nCapital : {capital}')
                    print('______________________________________________')

                elif ltp > buy:

                    if short_target_1_flag:

                        short_position_flag = False
                        capital = capital - (position_size/2 * ltp)
                        #telegram_bot.send(f'STOP LOSS - Price : {ltp} \nGameplan Sell Price : {buy}')
                        print(f'{index} \nSTOP LOSS - Price : {ltp} \nGameplan Buy Price : {buy} \nCapital : {capital}')
                        print('______________________________________________')

                    else:

                        short_position_flag = False
                        capital = capital - (position_size * ltp)
                        #telegram_bot.send(f'STOP LOSS - Price : {ltp} \nGameplan Sell Price : {buy}')
                        print(f'{index} \nSTOP LOSS - Price : {ltp} \nGameplan Buy Price : {buy} \nCapital : {capital}')
                        print('______________________________________________')

                    if cci > 20:

                        long_position_flag = True
                        capital = capital - (position_size * ltp)
                        #telegram_bot.send(f'LONG ENTRY - Price : {ltp} \nGameplan Buy Price : {buy}')
                        #print(f'{index} \nLONG ENTRY - Price : {ltp} \nGameplan Buy Price : {buy} \nCapital : {capital}')
                        print(f'{index} \nLONG ENTRY - Price : {ltp} \nGameplan Buy Price : {buy}')
                        print('______________________________________________')


            if (((long_position_flag | short_position_flag) == True) & (index.time() >= datetime.time(hour=15, minute=29))):

                if long_position_flag:
                    
                    if long_target_1_flag:

                        long_position_flag = False
                        capital = capital + (position_size/2 * ltp)
                        #telegram_bot.send(f'END OF DAY CLOSING LONG POSITION - Price : {ltp}')
                        print(f'{index} \nEND OF DAY CLOSING LONG POSITION - Price : {ltp} \nCapital : {capital}')
                        print('______________________________________________')

                    else:
                        
                        long_position_flag = False
                        capital = capital + (position_size * ltp)
                        #telegram_bot.send(f'END OF DAY CLOSING LONG POSITION - Price : {ltp}')
                        print(f'{index} \nEND OF DAY CLOSING LONG POSITION - Price : {ltp} \nCapital : {capital}')
                        print('______________________________________________')

                        
                if short_position_flag:

                    if short_target_1_flag:
                    
                        short_position_flag = False
                        capital = capital - (position_size/2 * ltp)
                        #telegram_bot.send(f'END OF DAY CLOSING SHORT POSITION- Price : {ltp}')
                        print(f'{index} \nEND OF DAY CLOSING SHORT POSITION - Price : {ltp} \nCapital : {capital}')
                        print('______________________________________________')
                        
                    else:
                        
                        short_position_flag = False
                        capital = capital - (position_size * ltp)
                        #telegram_bot.send(f'END OF DAY CLOSING SHORT POSITION- Price : {ltp}')
                        print(f'{index} \nEND OF DAY CLOSING SHORT POSITION - Price : {ltp} \nCapital : {capital}')
                        print('______________________________________________')
```
