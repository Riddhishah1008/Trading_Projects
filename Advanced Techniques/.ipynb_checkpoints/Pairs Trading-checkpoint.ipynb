```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import datetime

import glob

import warnings
warnings.filterwarnings('ignore')
```

```python
# Get paths to the NIFTY100 stock csv files
stock_data_paths = glob.glob('/Users/rshah181/Desktop/AMADEUS/NIFTY100/*')
```

```python
# Extract stock tickers from the stock_data_paths list
stock_tickers = list(pd.Series(stock_data_paths).apply(lambda x: x.split('/')[-1].split('.')[0]))
```

```python
# List to store tickers of stocks with insufficient data
failed_tickers = []
```

```python
# DataFrame to store closing prices of the stocks
master_price_data = pd.DataFrame()
```

```python
# Iterate over the tickers
for ticker in stock_tickers:
    
    # Read price data CSV of the selected ticker
    price_data = pd.read_csv('/Users/rshah181/Desktop/AMADEUS/NIFTY100/' + ticker + '.csv')

    # Check if the # of rows is less thatn 3500
    if len(price_data) < 3500:
        failed_tickers.append(ticker)
        continue

    # Clean up price data and set datetimec column as index
    price_data['datetime'] = price_data['datetime'].apply(lambda x: datetime.datetime.strptime(x.split()[0], '%Y-%m-%d'))
    price_data = price_data.set_index('datetime')
    price_data = price_data.dropna()

    # Store close prices in a variable called close
    close = price_data['close']
    
    # Resample close prices to daily timeframe to remove any timestamp duplicates
    close = close.resample('D').first()

    # Add close prices of the ticker to master_price_data
    master_price_data[ticker] = close
```

```python
# Replace 0s with NAs and drop rows with NAs
master_price_data = master_price_data.replace(0, np.nan).dropna()
```

```python
# Split into in sample, validation and out sample sets
in_sample = master_price_data.iloc[:int(len(master_price_data)/3)]
validation = master_price_data.iloc[int(len(master_price_data)/3):int(len(master_price_data)/2)]
out_sample = master_price_data.iloc[int(len(master_price_data)/2):]
```

### Correlation

**A number between -1 and 1**

$ \\ $

corr(X, Y) = $\dfrac{(\sum x_i - \bar{x})(\sum y_i - \bar{y})}{\sqrt{(\sum x_i - \bar{x})^2(\sum y_i - \bar{y})^2}}$


```python
# Compute correlations between the stocks & store in correlation table
corr_table = in_sample.corr()
```

```python
corr_table
```

```python
# Create dataframe to log the best correlation pairs
corr_log = pd.DataFrame(columns = ['X', 'Y', 'Corr'])
```

```python
# Iterate over the tickers/stocks
for ticker in in_sample.columns:
    
    # For the given ticker as X, select all stocks Y where 0.9 < Corr(X, Y) < 1 
    for x, y in zip([ticker]*len(corr_table[ticker][(corr_table[ticker] > 0.9) & (corr_table[ticker] < 1)].index), corr_table[ticker][(corr_table[ticker] > 0.9) & (corr_table[ticker] < 1)].index):
        
        # Add the X, Y pair to the corr_log
        corr_log.loc[len(corr_log)] = [x, y, round(corr_table[x][y], 2)]
```

```python
corr_log.sort_values('Corr', ascending=False)
```

```python
# Select one of the top pairs from the correlation log
x = corr_log.sort_values('Corr').iloc[-3]['X']
y = corr_log.sort_values('Corr').iloc[-3]['Y']
```

```python
# Store the X, Y, tickers in a tuple called pair
pair = (x, y)
print(pair)
```

```python
# Set matplotlib plot style
plt.style.use(plt.style.available[3])
plt.style.use('dark_background')
```

```python
# Plot cumulative returns of the pair
in_sample[[x, y]].pct_change().cumsum().plot(figsize=(10, 8));
```

```python
# Set figure size
plt.figure(figsize=(10, 7))
# Create linearly spaced array for the colors
colors = np.linspace(0.1, 1, len(in_sample))
# Create a scatter plot object
scatter = plt.scatter(in_sample[pair[0]], in_sample[pair[1]], s=30, c=colors, cmap=plt.get_cmap('cool'), edgecolor='k', alpha=0.5)

# Colorbar Stuff...
cb = plt.colorbar(scatter)
cb.ax.set_yticklabels([str(p.date()) for p in in_sample[::len(in_sample)//9].index])

# x and y labels
plt.xlabel(pair[0])
plt.ylabel(pair[1]);
```

### Linear Regression

**Drawing the Line of Best Fit**

$ \\ $

$ y = \beta_1X + \beta_0 + \varepsilon$

$\beta_1$ : Slope

$\beta_0$ : Intercept

$\varepsilon$ : Error

***

#### Linear Regression using statsmodels

```python
# Import statsmodels
import statsmodels
import statsmodels.api as sm
```

```python
# Fit OLS Regression to the pair
X = sm.add_constant(in_sample[pair[0]], prepend=False)
ols = sm.OLS(in_sample[pair[1]], X).fit()
beta = ols.params
```

```python
print(ols.summary2())
```

```python
beta
```

### price(BOB) = 0.728686*price(PNB) - 16.140944

```python
# Create Scatter Plot
plt.figure(figsize=(15, 10))
colors = np.linspace(0.1, 1, len(in_sample))
scatter = plt.scatter(in_sample[pair[0]], in_sample[pair[1]], s=30, c=colors, cmap=plt.get_cmap('cool'), edgecolor='k', alpha=0.4)

# Plot line of best fit
y_fit = [X.min().dot(beta), X.max().dot(beta)]
plt.plot([X.min()[0], X.max()[0]], y_fit, '--w', linewidth=3, label='Line of Best Fit')
plt.legend()

# Display Colorbar
cb = plt.colorbar(scatter)
cb.ax.set_yticklabels([str(p.date()) for p in in_sample[::len(in_sample)//9].index])

# Display x & y labels
plt.xlabel(pair[0])
plt.ylabel(pair[1]);
```

### Modelling the Spread

$ \varepsilon = y- \beta_1X - \beta_0$


```python
# Create the spread
spread = pd.DataFrame(in_sample[pair[1]] - beta[0] * in_sample[pair[0]] - beta[1])
spread.columns = ['in-sample']
```

```python
spread
```

```python
# Plot the Spread
spread.plot();
plt.xlabel('Time');
plt.ylabel('Spread');
```

#### Cointegration Test

```python
# Run the ADF test on the spread
adf = statsmodels.tsa.stattools.adfuller(spread['in-sample'], maxlag=1)
print('ADF test statistic: %s' % (adf[0]))
print('p-value: %s' % (np.round(adf[1], 4)))
```

```python
# Plot spread, mean and std dev
spread.plot(figsize=(15, 5));
plt.axhline(spread['in-sample'].mean(), ls = '--', color='r', lw=2);
plt.axhline(spread['in-sample'].mean() + spread['in-sample'].std()*2, ls = '--', color='y', lw=2);
plt.axhline(spread['in-sample'].mean() - spread['in-sample'].std()*2, ls = '--', color='y', lw=2);
```

#### Validating the results....

```python
# Create spread on the validation dataset
spread_val = spread.reindex((spread.index).union(validation.index))
spread_val['validation'] = validation[pair[1]] - beta[0] * validation[pair[0]] - beta[1]
```

```python
# Plot the validation spread
spread_val.plot(figsize=(15, 5));
plt.axhline(spread['in-sample'].mean(), ls = '--', color='r', lw=2);
plt.axhline(spread['in-sample'].mean() + spread['in-sample'].std(), ls = '--', color='y', lw=2);
plt.axhline(spread['in-sample'].mean() - spread['in-sample'].std(), ls = '--', color='y', lw=2);
```

```python
# Combine in-sample and validation sets into single dataframe
inandval = in_sample.append(validation)
```

## C4.6: Kalman Filter
***

```python
# Import KalmanFilter class from pykalman library
from pykalman import KalmanFilter
```

```python
# Create observation matrix
obs_mat = sm.add_constant(inandval[pair[0]].values, prepend=False)[:, np.newaxis]

# Create Kalman Filter object called kf
kf = KalmanFilter(n_dim_obs=1, n_dim_state=2, initial_state_mean=np.ones(2), initial_state_covariance=np.ones((2, 2)), transition_matrices=np.eye(2), observation_matrices=obs_mat, observation_covariance=10**2, transition_covariance=0.01**2 * np.eye(2))   
```

```python
# Run the Kalman filter
state_means, state_covs = kf.filter(inandval[pair[1]])
```

```python
# Store beta values in dataframe beta_kf
beta_kf = pd.DataFrame({'Slope': state_means[:, 0], 'Intercept': state_means[:, 1]}, index=inandval.index)
```

### y = mx + c

#### m -> slope
#### c -> intercept

```python
# Plot beta values computed by the Kalman filter
beta_kf.plot(subplots=True);
```

```python
# Create Scatter Plot
plt.figure(figsize=(15, 10))
colors = np.linspace(0.1, 1, len(inandval))
scatter = plt.scatter(inandval[pair[0]], inandval[pair[1]], s=30, c=colors, cmap=plt.get_cmap('cool'), edgecolor='k', alpha=0.5)

# Display Colorbar
cb = plt.colorbar(scatter)
cb.ax.set_yticklabels([str(p.date()) for p in inandval[::len(inandval)//10].index])

# Display x & y labels
plt.xlabel(pair[0])
plt.ylabel(pair[1]);

# Add kalman filter regression lines
step = 50
xi = np.linspace(inandval[pair[0]].min(), inandval[pair[0]].max(), 2)
colors_l = np.linspace(0.1, 1, len(state_means[::step]))
for i, b in enumerate(state_means[::step]):
    plt.plot(xi, b[0] * xi + b[1], alpha=.5, ls='-', c=plt.get_cmap('cool')(colors_l[i]))
```

```python
# Plot spread as per static beta values
spread_val.plot(style=['g', '--b', '--y', '--y', 'r'], figsize=(15, 5));
plt.axhline(spread['in-sample'].mean(), ls = '--', color='b');
plt.axhline(spread['in-sample'].mean() + spread['in-sample'].std(), ls = '--', color='y');
plt.axhline(spread['in-sample'].mean() - spread['in-sample'].std(), ls = '--', color='y');
plt.legend(loc=0);
```

### Trading Strategy

```python
# Get closing prices of the pair and store in a dataframe
pairs_data = inandval[[pair[0], pair[1]]].copy()
```

```python
# Add the spread as a new column
pairs_data['spread'] = spread_kf
```

```python
# Import RSI from pyti
from pyti.relative_strength_index import relative_strength_index as calc_rsi
```

```python
# Compute 20-period RSI values of the spread
pairs_data['rsi'] = calc_rsi(pairs_data['spread'], 20)

# Compute buy points and sell points based on the RSI
pairs_data['buy'] = pairs_data[pairs_data['rsi'] < 45]['spread']
pairs_data['sell'] = pairs_data[pairs_data['rsi'] > 55]['spread']
```

```python
# Plot the buy and sell points on the spread
pairs_data['spread'].iloc[-500:].plot(figsize=(15, 5))
plt.scatter(pairs_data.index, pairs_data['buy'], c='w', s=100)
plt.scatter(pairs_data.index, pairs_data['sell'], c='r', s=100)
```

```python
pairs_data
```

```python
# Convert buy and sell points into integers (1, 0)
pairs_data['buy']= (abs(pairs_data['buy']) > 0).apply(int)
pairs_data['sell'] = (abs(pairs_data['sell']) > 0).apply(int)

# Create a column naming the "perspective" ticker
pairs_data['who?'] = pair[1]
```

```python
pairs_data.tail(10)
```

```python
# Create a ratio column and store the slope
pairs_data['ratio'] = beta_kf['Slope'].shift()
```

```python
# Compute the close to next close changes, weigh by the ratio and shift back by one row
pairs_data[pair[0] + '_chg'] = np.round(pairs_data[pair[0]].diff().shift(-1)*pairs_data['ratio'], 2)
pairs_data[pair[1] + '_chg'] = pairs_data[pair[1]].diff().shift(-1)
```

```python
# Compute PnL of buying the spread
pairs_data['buy_pnl'] = (pairs_data['buy']*pairs_data[pair[1] + '_chg'] - pairs_data['buy']*pairs_data[pair[0] + '_chg'])   

# Compute PnL of selling the spread
pairs_data['sell_pnl'] = (- pairs_data['sell']*pairs_data[pair[1] + '_chg'] + pairs_data['sell']*pairs_data[pair[0] + '_chg'])    
```

```python
pairs_data.iloc[155:168]
```

```python
# Combine the buy and sell PnLs, fill NAs with 0s, sum along axis 1, replace 0s with NAs, drop NAs, round everthing to 2 decimal places
pairs_data['pnl'] = np.round(pairs_data[['buy_pnl', 'sell_pnl']].fillna(0).sum(axis=1).replace(0, np.nan), 2)
```

```python
# Extract pnl from the dataframe and store in new variable
pnl = pairs_data['pnl'].dropna()
```

```python
# Display Winrate % and R:R ratio
print('Win Rate % : ', np.round((sum(pnl > 0)/len(pnl))*100, 2))
print('Risk:Reward: ', np.round(abs(pnl[pnl > 0].mean()/pnl[pnl < 0].mean()), 2))
```

```python
# Plot beautiful cumulative PnL graph 
pnl.cumsum().plot()
```
